```python
import requests
from bs4 import BeautifulSoup

# URL de la pÃ¡gina a scrapear
url = "https://boardgamegeek.com/boardgame/295947/cascadia/wiki"

# Enviar una solicitud GET
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    html_content = response.text
    print("PÃ¡gina obtenida con Ã©xito")
else:
    print(f"Error {response.status_code}")
```

```python
# Crear el objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Imprimir el HTML formateado
print(soup.prettify()[:1000])  # Solo mostramos una parte para no saturar el output
```

```python
!pip install selenium webdriver-manager
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# Configurar opciones de Chrome (opcional)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # si no querÃ©s que abra ventana del navegador
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Iniciar el driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
url = "https://boardgamegeek.com/boardgame/295947/cascadia/wiki"
driver.get(url)
```

```python
print(driver.current_url)
```

```python
headers = driver.find_elements(By.XPATH, "//h1 | //h2 | //div")

for i, el in enumerate(headers):
    try:
        text = el.text.strip()
        if len(text) > 30:
            print(f"ðŸ”¹ Element #{i} - Tag: {el.tag_name} - Text: {text[:80]}...\n")
    except:
        continue
```

```python
from selenium.webdriver.common.by import By

# Obtener todos los divs visibles
divs = driver.find_elements(By.TAG_NAME, "div")

# Usamos un set para eliminar duplicados automÃ¡ticamente
contenidos_validos = set()

for i, div in enumerate(divs):
    try:
        text = div.text.strip()
        if text.startswith("Community Wiki Edit\nMicrobadges") and "COMPONENTS:" in text:
            if text not in contenidos_validos:
                print(f"âœ… Contenido vÃ¡lido Ãºnico en div #{i}\n{'-'*40}")
                print(text[:1000], "...\n")
                contenidos_validos.add(text)
    except:
        continue
```

```python
contenidos_validos = set()

for i, div in enumerate(divs):
    try:
        text = div.text.strip()
        if text.startswith("Community Wiki Edit\nMicrobadges") and "COMPONENTS:" in text:
            contenidos_validos.add(text)  # âœ… solo el texto, sin prints
    except:
        continue
```

```python
# Ruta de destino
ruta_salida = "/content/drive/MyDrive/NLP_CASCADIA/02_DATOS-EXTRAIDOS/Cascadia_CommunityWiki.txt"

# Escribir el contenido limpio
with open(ruta_salida, "w", encoding="utf-8") as f:
    for contenido in contenidos_validos:
        f.write(contenido + "\n\n")

print("âœ… Archivo guardado en:", ruta_salida)
```
