```python
def pdf_to_text(pdf_path, output_txt):
    # Open the PDF file in read-binary mode
    with open("/content/04_SOURCES/Cascadia-Rules_ES-web.pdf", 'rb') as pdf_file:
        # Create a PdfReader object instead of PdfFileReader
        pdf_reader = PyPDF2.PdfReader("/content/04_SOURCES/Cascadia-Rules_ES-web.pdf")

        # Initialize an empty string to store the text
        text = ''

        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()

    # Write the extracted text to a text file
    with open(output_txt, 'w', encoding='utf-8') as txt_file:
        txt_file.write(text)

if __name__ == "__main__":
    pdf_path = r'/content/04_SOURCES/Cascadia-Rules_ES-web.pdf'

    output_txt = r'reglamento.txt'

    pdf_to_text(pdf_path, output_txt)

    print("PDF converted to text successfully!")
```
