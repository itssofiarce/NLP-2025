```python
import requests
from bs4 import BeautifulSoup

# URL de la p√°gina a scrapear
url = "https://boardgamegeek.com/boardgame/295947/cascadia"

# Enviar una solicitud GET
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    html_content = response.text
    print("P√°gina obtenida con √©xito")
else:
    print(f"Error {response.status_code}")
```

```python
# Crear el objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Imprimir el HTML formateado
print(soup.prettify()[:1000])  # Solo mostramos una parte para no saturar el output
```

```python
# Busco el t√≠tulo
description_title = soup.find('h2', string='Description')
```

```python
# Busco el cuerpo de la descripcion
description_body = soup.find('div', class_='game-description')
if description_body:
    print(description_body.get_text(strip=True))
else:
    print("‚ùå No se encontr√≥ el cuerpo de la descripci√≥n.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")  # Ejecutar en modo sin ventana (ideal para Colab)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Iniciar el driver
driver = webdriver.Chrome(service=Service("/usr/local/bin/chromedriver"), options=chrome_options)

# Probar cargando la p√°gina
driver.get("https://boardgamegeek.com/boardgame/295947/cascadia")
print("‚úÖ P√°gina cargada correctamente:", driver.title)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Suponiendo que ya ten√©s el driver configurado y la p√°gina cargada:
driver.get("https://boardgamegeek.com/boardgame/295947/cascadia")

# Esperamos que la p√°gina cargue completamente
driver.implicitly_wait(10)
```

```python
# Extraer el t√≠tulo "Description"
description_title_text = ""

try:
    description_title = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//h3[contains(normalize-space(), 'Description')]"))
    )
    description_title_text = description_title.text.strip()
    print("‚úÖ T√≠tulo encontrado:", description_title_text)
except Exception as e:
    print("‚ùå No se encontr√≥ el t√≠tulo:", e)
```

```python
# Extraer el cuerpo de la descripci√≥n
description_body_text = ""

try:
    description_body = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, ".game-description-body"))
    )
    description_body_text = description_body.text
    print("\nüìù Descripci√≥n completa:\n")
    print(description_body_text)
except Exception as e:
    print(f"‚ùå No se encontr√≥ el cuerpo de la descripci√≥n: {e}")
```
