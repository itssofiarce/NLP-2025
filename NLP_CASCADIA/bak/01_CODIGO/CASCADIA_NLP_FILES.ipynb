```python
import fitz

# 2. Directorios
base_path = "/content/drive/MyDrive/NLP_CASCADIA"
source_folder = os.path.join(base_path, "04_SOURCES")
output_folder = os.path.join(base_path, "02_DATOS-EXTRAIDOS")

# 3. Archivos a procesar
nombres_archivos = [
    "FILE1_Cascadia_TeachingScript",
    "FILE2_Cascadia_QuickRefScoringWildlifeCards",
    "FILE3_Cascadia_CondensedRules"
]

# 4. Procesar cada PDF
for nombre in nombres_archivos:
    ruta_pdf = os.path.join(source_folder, f"{nombre}.pdf")
    ruta_txt = os.path.join(output_folder, f"{nombre}.txt")

    texto_extraido = ""

    try:
        with fitz.open(ruta_pdf) as doc:
            for page in doc:
                texto_extraido += page.get_text() + "\n"

        with open(ruta_txt, "w", encoding="utf-8") as f:
            f.write(texto_extraido)

        print(f"✅ Guardado: {ruta_txt}")

    except Exception as e:
        print(f"❌ Error con {nombre}: {str(e)}")
```

```python
!pip install python-docx
from docx import Document

# Ruta al archivo Word en Drive
docx_path = '/content/drive/MyDrive/NLP_CASCADIA/04_SOURCES/FILE4_Cascadia_Rules.docx'

# Ruta para guardar el .txt extraído
txt_path = '/content/drive/MyDrive/NLP_CASCADIA/02_DATOS-EXTRAIDOS/FILE4_Cascadia_Rules.txt'

try:
    # Abrir el archivo Word
    doc = Document(docx_path)

    # Extraer todo el texto
    texto_extraido = "\n".join([p.text for p in doc.paragraphs])

    # Guardar el texto en un archivo .txt
    with open(txt_path, 'w', encoding='utf-8') as f:
        f.write(texto_extraido)

    print(f"✅ Texto extraído y guardado en:\n{txt_path}")

except Exception as e:
    print(f"❌ Error al procesar el archivo Word: {e}")
```
