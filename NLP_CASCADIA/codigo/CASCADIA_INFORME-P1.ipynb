```python
# Instalar pdfminer.six si no estÃ¡ instalado
!pip install pdfminer.six

from pdfminer.high_level import extract_text
import requests
import os

# Descargar el PDF
pdf_url = "https://www.alderac.com/wp-content/uploads/2021/08/Cascadia-Rules.pdf"
response = requests.get(pdf_url)
with open("cascadia_manual.pdf", "wb") as f:
    f.write(response.content)

# Extraer texto
text = extract_text("cascadia_manual.pdf")

# Ruta a carpeta de Google Drive
output_folder = "/content/drive/MyDrive/NLP_CASCADIA"
output_filename = "cascadia_manual.txt"
output_path = os.path.join(output_folder, output_filename)

# Crear la carpeta si no existe
os.makedirs(output_folder, exist_ok=True)

# Guardar el texto del PDF
with open(output_path, "w", encoding="utf-8") as f:
    f.write(text)

print(f"\nðŸ’¾ Texto del PDF guardado en: {output_path}")
print("Â¿El archivo existe?", os.path.exists(output_path))
```

## Extraccion-2

```python
# Instalar BeautifulSoup si no estÃ¡ instalado
!pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup
import os
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive', force_remount=True)

# Carpeta base donde se guardarÃ¡n los textos
base_folder = "/content/drive/MyDrive/NLP_CASCADIA/02_WEB"
os.makedirs(base_folder, exist_ok=True)

def extract_text_from_url(url, filename):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Eliminar scripts y estilos
    for script in soup(["script", "style"]):
        script.decompose()

    text = soup.get_text(separator="\n")

    # Ruta final del archivo
    output_path = os.path.join(base_folder, filename)

    # Guardar el texto
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(text)

    print(f"âœ… Guardado: {output_path}")

# Diccionario de URLs y nombres de archivo
urls = {
    "https://www.flatout.games/cascadia": "flatout_games.txt",
    "https://www.whatboardgame.com/post/cascadia-board-game-review": "whatboardgame_review.txt",
    "https://boardgamereview.co.uk/game-reviews/cascadia-board-game-review/": "boardgamereview_review.txt",
    "https://oneboardfamily.com/cascadia-review/": "oneboardfamily_review.txt",
    "https://www.board-game.co.uk/how-to-play-cascadia/": "board_game_co_uk_guide.txt",
    "https://www.bluehighwaygames.com/cascadia.html": "bluehighwaygames_description.txt"
}

# Extraer texto de cada URL
for url, filename in urls.items():
    extract_text_from_url(url, filename)
```
